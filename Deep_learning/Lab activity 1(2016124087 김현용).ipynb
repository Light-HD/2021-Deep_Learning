### **Linear regression using only Numpy**

```python
import numpy as np

# To plot pretty figures
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt

# np.random.rand
# Create an array of the given shape and populate it with random samples from a uniform distribution over [0, 1)
X = 3 * np.random.rand(100, 1)
y = 5 + 6 * X + np.random.randn(100, 1)

plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 3, 0, 30])
plt.show()

# adding a "dummpy" input x0 for the bias
X_b = np.c_[np.ones((100, 1)), X]  # add x0 = 1 to each instance 
#print(X_b)
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y) # The Normal Equation: (X^T * X)^-1 * X^T * y

test_npc = np.c_[np.array([1,2,3]), np.array([4,5,6])]

np.c_[np.array([[1,2,3]]), 0, 0, np.array([[4,5,6]])]

X_new = np.array([[0], [3]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)

plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0, 3, 0, 30])
plt.show()
```

### **Linear regression using Scikit-Learn**

```python
# Scikit-Learn ≥0.20 is required
import sklearn
assert sklearn.__version__ >= "0.20"
from sklearn.linear_model import LinearRegression

X = 3 * np.random.rand(100, 1)
y = 5 + 6 * X + np.random.randn(100, 1)

plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 3, 0, 30])
plt.show()

lin_reg = LinearRegression()
lin_reg.fit(X, y)

#Here
# lin_reg.intercept_ => 'bias'
# lin_reg.coef_ => weights
lin_reg.intercept_, lin_reg.coef_

y_predict = lin_reg.predict(X_new)

plt.plot(X_new, y_predict, "r-", linewidth=2, label="Predictions")
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([0, 3, 0, 30])
plt.show()
```
